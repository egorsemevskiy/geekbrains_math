Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import math
```

```python
def length(x):
    sqarearray = 0
    for r in x:
        sqarearray += r**2 
    return math.sqrt(sqarearray)
```

```python
length([10,5,4,5,10])
```

```python
def xy(r,phi):
    return r*np.cos(phi), r*np.sin(phi)

fig = plt.figure()
ax = fig.add_subplot(111,aspect='equal')

phis=np.arange(0,6.28,0.01)
r =1.
ax.plot( *xy(r,phis), c='r',ls='-' )
plt.show()
```

### 5. Задание (в программе)

Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from mpl_toolkits.mplot3d import axes3d
ax = axes3d.Axes3D(plt.figure())

i = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(i, i)
Z = X + Y
Z2 = X + Y + 2

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.plot_wireframe(X, Y, Z2, rstride=10, cstride=10)

plt.show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from mpl_toolkits.mplot3d import axes3d
ax = axes3d.Axes3D(plt.figure())

i = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(i, i)
Z = X**2 - Y
Z2 = X**2 + Y**2 + 2

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.plot_wireframe(X, Y, Z2, rstride=10, cstride=10)

plt.show()
```
