### 3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.


```python
 
def dectopol(x, y):
    r = np.sqrt(x**2 + y**2)
    p = np.degrees(np.arctan2(y, x))
    return r, p

i = np.array([12])
q = np.array([13])
 
print(dectopol(i, q))
```

Напишите код, который будет рисовать график окружности в полярных координатах.

```python
fig = plt.figure(figsize=(10., 10.))
 
phi = np.arange(0,2*np.pi,0.01)
rho = 5 * np.cos(phi)
 
ax1 = fig.add_subplot(232, projection='polar')
ax1.plot(rho, phi)
ax1.grid(True)
 
plt.show()
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
fig = plt.figure(figsize=(10., 10.))

phi = np.arange(0,2*np.pi,0.01)
rho = 6 / (2 - np.cos(phi))

ax1 = fig.add_subplot(232, projection='polar')
ax1.plot(rho, phi)
ax1.grid(True)
 
plt.show()
```

### 4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x = np.arange(-5, 5, .01)
y1 = x ** 2 - 1
y2 = 1 - (1 - np.exp(x)) / x 
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='green')
plt.show()
```

```python
def func(x):
    return x ** 2 - 1 - (1 - (1 - np.exp(x)) / x)

from scipy.optimize import fsolve

print(f'Первый корень уравнения равен {(fsolve(func, -2))}')
print(f'Второй корень уравнения равен {(fsolve(func, 2))}')
print(f'Третий корень уравнения равен {(fsolve(func, 5))}')
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1

```python
def func1(x):
    return (x ** 2 - 1 - (1 - (1 - np.exp(x)) / x)) > 0
y = []
for i in x:
    if func1(i):
        y.append(i ** 2 - 1)
    else:
        y.append(None)
        
plt.plot(x, y)
plt.show()
```
